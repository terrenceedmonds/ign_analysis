Okay, want to create some functions to get rid of some of this unicode

```python
def text_cleaner(text):

    final_text = text[40:]
    final_text = re.sub('\\\\u003e', '', final_text)
    final_text = re.sub('\\\\u003cbr /', '', final_text)
    final_text = re.sub('\\\\u003csection', '', final_text)
    final_text = re.sub('\\\\u003c/section', '', final_text)
    final_text = re.sub('\\\\u003cp', '', final_text)
    final_text = re.sub('\\\\u003cstrong', '', final_text)
    final_text = re.sub('\\\\u003c/strong', '', final_text)
    final_text = re.sub('\\\\u003c/p', '', final_text)
    final_text = re.sub('data-transform', '', final_text)
    final_text = re.sub('=', '', final_text)
    final_text = re.sub('cellpadding', '', final_text)
    final_text = re.sub('cellspacing', '', final_text)
    final_text = re.sub('table width', '', final_text)
#     final_text = re.sub('\\\\"', '', final_text)
#     final_text = re.sub('/section', '', final_text)

    # consider getting rid of the later half of the string, as it largely duplicates the review

    rough_cut = re.search('paginatedH', final_text)
    final_text = final_text[:rough_cut.start()]
    
    start_iter = re.finditer('start image div', final_text)
    end_iter = re.finditer('end image div', final_text)
    image_starts = []
    image_ends = []
    
    for iter1 in start_iter:
        image_starts.append(iter1.start() )
    for iter2 in end_iter:
        image_ends.append(iter2.end() )

    for i in range (0, len(image_starts) ):
        if i == 0:
            mid_text = final_text[:image_starts[i]] 
            
        elif i >= 1 and i < len(image_starts)-1:
            mid_text += final_text[image_ends[i-1]:image_starts[i]]
            
        elif i == len(image_starts)-1:
            mid_text += final_text[image_ends[i-1]:image_starts[i]] + final_text[image_ends[i]:]
    
    final_text = mid_text
    
    # commented out because need to examine the href a bit more
#     href_starts = []
#     href_ends = []
#     href_start_iter = re.finditer('href', final_text)
#     href_end_iter = re.finditer('html', final_text)
    
#     for iter1 in href_start_iter:
#         href_starts.append(iter1.start() )
#     for iter2 in href_end_iter:
#         href_ends.append(iter2.end() )
        
#     print(len(href_starts))
#     print(len(href_ends))
#     for i in range(0, len(href_starts) ):
#         if i == 0:
#             mid_text = final_text[:href_starts[i]]
            
#         elif i >= 1 and i < len(href_starts)-1:
#             mid_text += final_text[href_ends[i-1]:href_starts[i]]
            
#         elif i == len(href_starts)-1:
#             mid_text += final_text[href_ends[i-1]:href_starts[i]] + final_text[href_ends[i]:]
            
#     final_text = mid_text
    
    return final_text

ignore_text_cut = review_script.string[mark1.end():]
print(text_cleaner(ignore_text_cut))
```

Going to crawl and download the IGN data with only some rough cleaning applied

```python
directory = 'ign_sitemap_data/'
files = os.listdir(directory)
paths = [directory + x for x in files if not re.search('.gz', x)]

article_links = []
# grabbing all links
for i in range(0, len(paths)):
    print(i, paths[i])
    if not re.search('.DS_Store', paths[i]):
        fopen = open(paths[i], encoding='utf-8')
        fread = fopen.read()
        file_soup = BeautifulSoup(fread)
        
        for loc in file_soup.find_all('loc'):
            if re.search('review', loc.string) and not re.search('preview', loc.string):
                article_links.append(loc.string)

# crawl all of the links

write_path = 'ign_data/unedited_review_html.csv'

print(len(article_links))

i = 0

links_to_ignore = []
if os.path.exists(write_path):
    file_exists = True
    ignore_df = pd.read_csv(write_path, sep='`')
    for rows in ignore_df.itertuples():
        links_to_ignore.append(rows.link)
else:
    file_exists = False

ordered_links = []
ordered_scripts = []
temp_dict = {}

for link in article_links:
    
    if file_exists and link in links_to_ignore:
        i += 1
        continue
    
    if i % 100 == 0 and len(ordered_links)>0: 
        print(i)
        temp_dict = {'link': ordered_links, 'some_unfiltered_html': ordered_scripts}
        write_df = pd.DataFrame.from_dict(temp_dict)
        ordered_scripts = []
        ordered_links = []
        if not file_exists:
            write_df.to_csv(path_or_buf = write_path, sep = '`', mode = 'w')
            file_exists = True
        else:
            write_df.to_csv(path_or_buf = write_path, sep = '`', mode = 'a', header = False)
        
    link_sess = sesh.get(link)
    review_soup = BeautifulSoup(link_sess.html.html)
    review_script = review_soup.find('script',{'id':'__NEXT_DATA__'})
    ordered_scripts.append(review_script)
    ordered_links.append(link)
#     temp_dict[str(i)] = {'link': link, 'some_unfiltered_html': review_script}
#     write_df = pd.DataFrame.from_dict(temp_dict[str(i)])
#     if not file_exists:
#         write_df.to_csv(path_or_buf = write_path, sep = '`', mode = 'w')
#         file_exists = True
#     else:
#         write_df.to_csv(path_or_buf = write_path, sep = '`', mode = 'a', header = False)    
    time.sleep(0.5)
    
    i += 1
#     if i > 4:
#         break
temp_dict = {'link': ordered_links, 'unfiltered_html': ordered_scripts}
write_df = pd.DataFrame.from_dict(temp_dict)
write_df.to_csv(path_or_buf=write_path, sep='`', mode = 'a', header = False)
print('completed')
```
